## Investigating variation among drugs of a given "therapeutic class" 
Goal: identify sources variation across states of drugs that are prescribed for similar therapeutic purposes. I use the same matrix, $G_{d,s}$ containing the deviance residual value of each drug $d$ in each state $s$. This is sort of a corrected measure of how high- or low- prescribed a drug is, as compared to the nationwide rate.

I joined the FDA database containing drug **established pharmacologic class (EPC)**, such as Tricyclics, SSRIs, etc.  I also got the Medicaid database containing cost information and generic availability for the drugs.

### Focus on  antidepressants
Looking at just at Antidepressants, I took the subset of $G_{d,s}$ where $d \in Antidepressants$. Then I normalized each state (column) to have the same mean and var, creating the matrix $G'_{d,s}$, containing only the drugs in this therapeutic class.  This was in order to remove the overall trend of antidepressant use across states.

Below, I plot a clustered heatmap of $G'_{d,s}$, clustered only by distance between the rows and columns of this matrix. Then, I label the Pharmacologic class of each drug (box titled "PHARM_CLASSES_EPC"). There appears to be some relationship between the clustering of the rows and tricyclics vs SSRIs

Then, I averaged together the values of $G'_{d,s}$ for the same EPC below.  The central/western states seem least likely to take Tricyclic and more likely to take Seratonin and Norepinephrine Reuptake Inhibitors.

## Does drug generic availability associate with patterns of use?
I compared drugs that are mostly generic vs mostly brand, in terms of how much state usage varied. Generics appear to have greater variation in use (or vice versa?).  Spearman cor between fraction generic and state variation was correlation=0.14, pvalue=2.147e-06.  The rank-sum test between mostly generic and mostly brand drugs pvalue was 4.6e-10.
