## How much of an effect do states have on drug?
I attempted to measure this by comparing counties to other counties that share a border (at any point).  For any county that lies along the state border, I gather demographic and drug data for each other county (both in-state and out-of-state) that it borders.  Then I compare the set of bordering county pairs to each other in terms of distance between each pair of drug vectors, or in terms of demographics vectors. 

I z-scored the drug distances within each state (y-axis below). For each county pair that was between two states, I averaged the z-score of that pair from both states' distributions.  This should tell you if a pair of counties is more or less distant from each other as compared to other county pairs from the same state.
